```python
df.hist(figsize=(12,10))
plt.show()
```

These histograms visualize the distribution of each feature in the dataset. They can help identify skewness, outliers, and the general shape of the data for each variable.

```python
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(), annot=True)
plt.show()
```

This heatmap shows the correlation coefficients between all pairs of numerical features in the dataset. Values close to 1 or -1 indicate strong positive or negative linear relationships, respectively, while values close to 0 indicate weak or no linear relationship.

```python
sns.scatterplot(x="Glucose", y="BMI", hue="Outcome", data=df)
plt.show()
```
