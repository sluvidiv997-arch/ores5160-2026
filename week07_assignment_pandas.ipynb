```python
import pandas as pd

# The file downloaded from the link is an Excel file (.xlsx)
# The name of the uploaded Excel file from `files.upload()` is 'us_air_quality_one_week_2022_06_23 (1).xlsx'
# Load the first 10,000 rows into a Pandas dataframe called airquality
airquality = pd.read_excel("us_air_quality_one_week_2022_06_23 (1).xlsx", nrows=10000)

# Display the first 5 rows of the dataframe
airquality.head()
```

Q2. Data Summary and Cleaning

	1.	Use .info() and .describe(include="all") to explore the dataset.\\

	2.	Identify any columns that have missing (NaN) values.\\

	3.	Drop all rows that contain missing values.\\

```python
# Use .info() to get a concise summary of the DataFrame
print('DataFrame Info:')
airquality.info()

print('\n' + '='*50 + '\n')

# Use .describe(include='all') to generate descriptive statistics
print('DataFrame Description (including all columns):')
airquality.describe(include='all')
```

```python
# Identify columns that have missing (NaN) values
print('Missing values per column:')
missing_values = airquality.isnull().sum()
# Filter to show only columns with missing values
columns_with_missing = missing_values[missing_values > 0]
if not columns_with_missing.empty:
    print(columns_with_missing)
else:
    print('No missing values found in any column.')
```

```python
# Drop all rows that contain missing values
initial_rows = airquality.shape[0]
airquality.dropna(inplace=True)
final_rows = airquality.shape[0]

print(f"Initial number of rows: {initial_rows}")
print(f"Number of rows after dropping missing values: {final_rows}")
print(f"Number of rows dropped: {initial_rows - final_rows}")

# Verify no more missing values
print('\nVerifying no more missing values:')
print(airquality.isnull().sum().sum()) # Should be 0 if all rows with NaNs are dropped
```

Q3. Data Visualization

	1.	Create a histogram of the pm25 values.\\

	2.	Create a boxplot for the pm25 column.\\


```python
import matplotlib.pyplot as plt
import seaborn as sns

# Filter for 'bc' pollutant
bc_data = airquality[airquality['pollutant'] == 'bc']

# Check if there is any 'bc' data
if not bc_data.empty:
    # 1. Create a histogram of the 'bc' values (from the 'value' column)
    plt.figure(figsize=(10, 6))
    sns.histplot(bc_data['value'], kde=True, bins=30)
    plt.title('Histogram of Black Carbon (BC) Values')
    plt.xlabel('BC Value')
    plt.ylabel('Frequency')
    plt.show()

    # 2. Create a boxplot for the 'bc' column (from the 'value' column)
    plt.figure(figsize=(8, 6))
    sns.boxplot(y=bc_data['value'])
    plt.title('Boxplot of Black Carbon (BC) Values')
    plt.ylabel('BC Value')
    plt.show()
else:
    print("No 'bc' data found in the 'pollutant' column. Unique pollutants are:")
    print(airquality['pollutant'].unique())
```
